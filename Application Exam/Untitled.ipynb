```python
def project_cleaner(proj):
    if proj[:2]=="op":
        return "opsandadmin"
    elif proj[:4]=="cult":
        return "cultureandmanagement"
    elif proj[:4]=="hirn":
        return "hiring"
    else:
        return proj

ci_df.project = ci_df.project.apply(lambda x: project_cleaner(x))
```
