# Finding insights on Project Hours

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime

%matplotlib inline
```

```python
# Read cleaned csv file into a pandas dataframe
ci_df = pd.read_csv("dailycheckins-cleaned-final.csv")
ci_df = ci_df.drop(columns=['Unnamed: 0'])
ci_df
```

```python
projs = ci_df.project.unique()

print(projs)
print(len(projs))
```

```python
datenow = datetime.datetime(2018,11,29)
dstart = datetime.datetime(2017,6,1)
```

```python
for proj in projs:
    projdata = ci_df[ci_df.project == proj]
    projdata.timestamp = pd.to_datetime(projdata.timestamp)
    plt.figure(figsize=(32,9))
    plt.title('Logs for the Project: '+ proj)
    plt.xlim(dstart, datenow) 
    plt.xlabel('Date')
    plt.ylabel('Hours')
    plt.bar(projdata.timestamp,projdata.hours)
```

```python
p = []
n = []

for proj in projs:
    projlog = ci_df[ci_df.project == proj]
    projlog = projlog.reset_index()
    useruse = projlog.user.unique()
    p.append(proj)
    n.append(int(len(useruse)))
    
plt.figure(figsize=(10,20))
plt.barh(p,n)
plt.xlabel('Number of Users')
plt.ylabel('Project')
plt.gca().invert_yaxis()
```
