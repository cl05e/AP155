```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np

%matplotlib inline
```

```python
# Read csv file into a pandas dataframe
ci = pd.read_csv("dailycheckins.csv")
ci
```

The dataframe has few errors, namely: inconsistent timestamp formats, Russian months in the dates, spelling errors, and null values. We first correct those errors before cleaning the data further.

After cleaning typographical errors in Excel, we feed it to a pandas dataframe to further adjust components such as the 'timestamp' column.

```python
# Read semi-cleaned csv file into a pandas dataframe
ci_df = pd.read_csv("dailycheckins-cleaned.csv")
ci_df
```

We check if there are null values.

```python
# Check for missing values
print('Are there missing values?',ci_df.isnull().values.any())
print(ci_df.isnull().sum())
```

```python
ci_df[ci_df.user.isna()]
```

We fill these values with 'person' for now. We will get back to these later.

```python
ci_df.user.fillna('person',inplace=True)

ci_df
```

Notice that there are Russian months in the timestamps. We change those to English.

```python
# Change Russian months to English
ci_df.timestamp = ci_df.timestamp.str.replace('января','January')
ci_df.timestamp = ci_df.timestamp.str.replace('февраля','February')
ci_df.timestamp = ci_df.timestamp.str.replace('марта','March')
ci_df.timestamp = ci_df.timestamp.str.replace('апреля','April')
ci_df.timestamp = ci_df.timestamp.str.replace('мая','May')
ci_df.timestamp = ci_df.timestamp.str.replace('июня','June')
ci_df.timestamp = ci_df.timestamp.str.replace('июля','July')
ci_df.timestamp = ci_df.timestamp.str.replace('августа','August')
ci_df.timestamp = ci_df.timestamp.str.replace('сентября','September')
ci_df.timestamp = ci_df.timestamp.str.replace('октября','October')
ci_df.timestamp = ci_df.timestamp.str.replace('ноября','November')
ci_df.timestamp = ci_df.timestamp.str.replace('декабря','December')

ci_df
```

To change the timezone, we convert the 'timestamp' column into a datetime index.

```python
ci_df.timestamp = pd.to_datetime(ci_df.timestamp)
ci_df
```

We find out that the timezones cannot be converted as there are inconsistencies with the format. We change that by adding a '+00:00' to each value in the 'timestamp' column.

```python
ci_df.timestamp = ci_df.timestamp.astype(str) + '+00:00'
ci_df
```

```python
ci_df = ci_df.set_index('timestamp')
ci_df.index = pd.to_datetime(ci_df.index)

ci_df.index
```

We have successfully converted the 'timestamp' column into a DateTime index. Now we change the timezone.

```python
ci_df.index = ci_df.index.tz_convert('Asia/Manila')

ci_df.index
```

.We reset the index in order for us to check if there are duplicates

```python
ci_df = ci_df.reset_index()
ci_df
```

```python
# Select duplicate rows except last occurrence based on all columns
duplicateRowsDF = ci_df[ci_df.duplicated(keep='last')]

print("Duplicate Rows except last occurrence based on all columns are :", duplicateRowsDF)
```

AHA! There are duplicte rows, as seen above. We keep the last one since those are the earliest logs for the duplicates.

```python
ci_df = ci_df.drop_duplicates(subset=None, keep='last', inplace=False)
ci_df
```

We reset the index so that the elements are properly numbered.

```python
ci_df = ci_df.reset_index()
```

```python
ci_df = ci_df.drop(columns=['index'])
ci_df
```

We change the format of the datetime for the dataset to be more consistent.

```python
ci_df.timestamp = ci_df.timestamp.dt.strftime('%Y/%m/%d %H:%M')
ci_df
```

```python
ci_df.to_csv(r'dailycheckins-cleaned-final.csv')
```

## Total Hours Logged

```python
c1=ci_df
userssum = c1.groupby('user').agg({'hours':sum})
userssum = userssum.drop(index='person')
userssum
```

```python
plt.figure(figsize=(10,10))
plt.barh(userssum.index,userssum.hours,0.5)
plt.xlabel('Hours Logged (from 2017-2019)')
plt.ylabel('User')
plt.gca().invert_yaxis()
```

People with more hours logged could be the founders, managers or directors of the company; while people with less hours logged might be interns, temporary employees, or newcomers.
