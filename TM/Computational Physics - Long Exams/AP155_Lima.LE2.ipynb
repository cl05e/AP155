### Problem 2

Assuming the voltages at the points labeled 1, 2 and 3 are of the form $V_1 = x_1e^{iωt}$, $V_2 = x_2e^{iωt}$, and $V_3 = x_3e^{iωt}$, applying Kirchhoff’s law at each of the three points, along with Ohm’s law and the capacitor law:  

$$ \tag{1a} \frac{V_1 - V_+}{R_1} + \frac{V_1 - 0}{R_4} + c_1\frac{dV_1}{dt} - c_1\frac{dV_2}{dt} = 0 $$  
$$ \tag{1b} \frac{V_2 - V_+}{R_2} + \frac{V_2 - 0}{R_5} + c_1\frac{dV_2}{dt} + c_2\frac{dV_2}{dt} - c_1\frac{dV_1}{dt} - c_2\frac{dV_3}{dt}= 0 $$  
$$ \tag{1c} \frac{V_v - V_+}{R_3} + \frac{V_3 - 0}{R_6} + c_2\frac{dV_3}{dt} - c_2\frac{dV_2}{dt} = 0 $$   

Substituting the values and the derivatives of the voltages:  

$$ \tag{2a} \frac{x_1e^{iωt} - x_+e^{iωt}}{R_1} + \frac{x_1e^{iωt} - 0}{R_4} + c_1iωx_1e^{iωt} - c_1iωx_2e^{iωt} = 0 $$  
$$ \tag{2b} \frac{x_2e^{iωt} - x_+e^{iωt}}{R_2} + \frac{x_2e^{iωt} - 0}{R_5} + c_1iωx_2e^{iωt} + c_2iωx_2e^{iωt} - c_1iωx_1e^{iωt} - c_2iωx_3e^{iωt}= 0 $$  
$$ \tag{2c} \frac{x_3e^{iωt} - x_+e^{iωt}}{R_3} + \frac{x_3e^{iωt} - 0}{R_6} + c_2iωx_3e^{iωt} - c_2iωx_2e^{iωt} = 0 $$   

Dividing the equations by e^{iωt}:  

$$ \tag{3a} \frac{x_1 - x_+}{R_1} + \frac{x_1 - 0}{R_4} + c_1iωx_1 - c_1iωx_2 = 0 $$  
$$ \tag{3b} \frac{x_2 - x_+}{R_2} + \frac{x_2 - 0}{R_5} + c_1iωx_2 + c_2iωx_2 - c_1iωx_1 - c_2iωx_3 = 0 $$  
$$ \tag{3c} \frac{x_3 - x_+}{R_3} + \frac{x_3 - 0}{R_6} + c_2iωx_3 - c_2iωx_2 = 0 $$   

Further simplifying the equations:  

$$ \tag{3a} \frac{x_1 - x_+}{R_1} + \frac{x_1}{R_4} + c_1iωx_1 - c_1iωx_2 = 0 $$  
$$ \tag{3b} \frac{x_2 - x_+}{R_2} + \frac{x_2}{R_5} + c_1iωx_2 + c_2iωx_2 - c_1iωx_1 - c_2iωx_3 = 0 $$  
$$ \tag{3c} \frac{x_3 - x_+}{R_3} + \frac{x_3}{R_6} + c_2iωx_3 - c_2iωx_2 = 0 $$   

Adding the terms with $x_+$ to all equations:  

$$ \tag{4a} \frac{x_1}{R_1} + \frac{x_1}{R_4} + c_1iωx_1 - c_1iωx_2 = \frac{x_+}{R_1} $$  
$$ \tag{4b} \frac{x_2}{R_2} + \frac{x_2}{R_5} + c_1iωx_2 + c_2iωx_2 - c_1iωx_1 - c_2iωx_3 = \frac{x_+}{R_2} $$  
$$ \tag{4c} \frac{x_3}{R_3} + \frac{x_3}{R_6} + c_2iωx_3 - c_2iωx_2 = \frac{x_+}{R_3} $$   

Factoring $x_1, x_2 $ and $x_3$ from the equations:  

$$ \tag{5a} \left(\frac{1}{R_1} + \frac{1}{R_4} + iωc_1\right)x_1 - iωc_1x_2 = \frac{x_+}{R_1} $$  
$$ \tag{5b} - iωc_1x_1 + \left(\frac{1}{R_2} + \frac{1}{R_5} + iωc_1 + iωc_2\right)x_2  - iωc_2x_3 = \frac{x_+}{R_2} $$  
$$ \tag{5c} - iωc_2x_2 + \left(\frac{1}{R_3} + \frac{1}{R_6} + iωc_2\right)x_3  = \frac{x_+}{R_3} $$   

Thus we have shown that the constants satisfy the equations 5a, 5b and 5c.

With R1 = R3 = R5 = 1 kΩ,  
R2 = R4 = R6 = 2 kΩ,  
C1 = 1 µF, C2 = 0.5 µF,  
x+ = 3 V, ω = 1000 s^−1  

Equations 5a, 5b and 5c become:  

$$\tag{6a} (0.0015+0.001i)x_1 - 0.001ix_2 = 0.003$$  
$$\tag{6b} -0.001ix_1 + (0.0015+0.0015i)x_2 - 0.0005ix_3 = 0.0015$$  
$$\tag{6c} - 0.0005ix_2 + (0.0015+0.0005i)x_3 = 0.003$$  

These equations can be represented through matrices:  

$$ \left( \begin{array}{ccc}
0.0015+0.001i & -0.001i & 0 \\
-0.001i & 0.0015+0.0015i & -0.0005i \\
0 & -0.0005i & 0.0015+0.0005i \end{array} \right)
\left( \begin{array}{ccc}
x_1 \\
x_2 \\
x_3 \end{array} \right) = 
\left( \begin{array}{ccc}
0.003 \\
0.0015 \\
0.003 \end{array} \right)$$

```python
from numpy import array
from numpy.linalg import solve
from cmath import polar, pi
#Define the constants
R1 = R3 = R5 = 1000.
R2 = R4 = R6 = 2000.
C1 = 1e-6
C2 = 0.5e-6
xplus = 3.
omega = 1000.

#Set up the arrays for the equation solver
A = array( \
         [[ 1/R1 + 1/R4 + 1j*omega*C1, -1j*omega*C1, 0 ], \
         [ -1j*omega*C1, 1/R2 + 1/R5 + 1j*omega*C1 +1j*omega*C2, -1j*omega*C2 ],
         [ 0, -1j*omega*C2, 1/R3 + 1/R6 + 1j*omega*C2 ]], complex)
v = xplus*array([1/R1, 1/R2, 1/R3], float)

#Calculate the solution
x = solve(A, v)

#Convert to polars and print
r, phi = polar(x[0])
print('V1 amplitude = {0:.2f}, phase = {1:.2f}'.format(r, phi*180/pi))
r, phi = polar(x[1])
print('V2 amplitude = {0:.2f}, phase = {1:.2f}'.format(r, phi*180/pi))
r, phi = polar(x[2])
print('V3 amplitude = {0:.2f}, phase = {1:.2f}'.format(r, phi*180/pi))
```

### Problem 4

$$\tag{1} \langle E \rangle = \int_{-\infty}^{\infty} \Psi^*\left(\frac{-\hbar^2}{2m}\frac{\partial^2}{\partial x^2}+V(x)\right)\Psi dx$$  
$$\tag{2} \langle E \rangle = \int_{-\infty}^{\infty} \frac{1}{\sqrt[4]{\pi \sigma^2}} e^{-x^2/2\sigma^2} \left[ \frac{-\hbar^2}{2m} \left[ \frac{1}{\sqrt[4]{\pi \sigma^2}} \left( \frac{-e^{-x^2/2\sigma^2}}{\sigma^2} + \frac{x^2}{\sigma^4}e^{-x^2/2\sigma^2}\right) \right] + V(x)\frac{1}{\sqrt[4]{\pi \sigma^2}}e^{-x^2/2\sigma^2} \right] dx$$  
$$\tag{3} \langle E \rangle = \frac{1}{\sqrt{\pi \sigma^2}} \int_{-\infty}^{\infty}  e^{-x^2/2\sigma^2} \left[ \frac{\hbar^2}{2m\sigma^2} \left(e^{-x^2/2\sigma^2} - \frac{x^2}{\sigma^2} e^{-x^2/2\sigma^2}\right) + V(x)e^{-x^2/2\sigma^2} \right] dx$$  
$$\tag{4} \langle E \rangle = \frac{1}{\sqrt{\pi \sigma^2}} \int_{-\infty}^{\infty}  \left[ \frac{\hbar^2}{2m\sigma^2} \left(1 - \frac{x^2}{\sigma^2}\right) + V(x) \right]e^{-x^2/\sigma^2} dx$$  

```python
from numpy import pi, tan, exp, cos, sqrt, linspace
from gaussxw import gaussxwab

eV_to_J = 1.60218e-19
hbar = 1.0545718e-34 # in (m^2)(kg/s)
m = 9.10938356e-31 # in kg
def f(x, sigma, V):
    cons = hbar**2/(2*m*sigma**2)
    return (cons*(1-x**2/sigma**2) + V)*exp(-x**2/sigma**2)

def V(x, V0=10., x0=1e-10):
    return V0*eV_to_J*(x/x0)**2 #from eV to J  

def solve_energy(V, sigma=1e-7):
    a = -pi/2
    b = pi/2
    N = 100
    z, w = gaussxwab(N, a, b)
    E = 0.
    for k in range(N):
        E += w[k]*f(sigma*tan(z[k]), sigma, V(sigma*tan(z[k])))/(cos(z[k])**2/sigma)
    return(E/sqrt(pi*sigma**2))

def solve_energy(V, sigma=1e-7):
    a = -pi/2
    b = pi/2
    N = 100
    z, w = gaussxwab(N, a, b)
    E = 0.
    for k in range(N):
        E += w[k]*f(sigma*tan(z[k]), sigma, V(sigma*tan(z[k])))/(cos(z[k])**2/sigma)
    return(E/sqrt(pi*sigma**2))

sigma_min = 1e-11
sigma_max = 1e-9
sigma = linspace(sigma_min, sigma_max, 100)
E = solve_energy(V, sigma)
plot(sigma, E)
```

```python
def golden_ratio(solve_energy, V, x1, x4):
    # Constants
    accuracy = 1e-15         # Required accuracy in m
    z = (1+sqrt(5))/2       # Golden ratio

    # Initial positions of the four points
    x1 = sigma_min
    x4 = sigma_max
    x2 = x4 - (x4-x1)/z
    x3 = x1 + (x4-x1)/z

    # Initial values of the function at the four points
    f1 = solve_energy(V, x1)
    f2 = solve_energy(V, x2)
    f3 = solve_energy(V, x3)
    f4 = solve_energy(V, x4)

    # Main loop of the search process
    while x4-x1>accuracy:
        if f2<f3:
            x4,f4 = x3,f3
            x3,f3 = x2,f2
            x2 = x4 - (x4-x1)/z
            f2 = solve_energy(V, x2)
        else:
            x1,f1 = x2,f2
            x2,f2 = x3,f3
            x3 = x1 + (x4-x1)/z
            f3 = solve_energy(V, x3)

    sigma_min_E = 0.5*(x1+x4)
    return sigma_min_E
```

```python
sigma_min_E = golden_ratio(solve_energy, V, sigma_min, sigma_max)
print("The minimum energy is found at sigma = {0:.6} m.".format(sigma_min_E))
print("This corresponds to a minimum energy of {0:.6} eV.".format(solve_energy(V, sigma_min_E)/eV_to_J))
```

The exact ground state energy $E_0$ as given in Eq. (3) is $E_0 = \hbar/x_0 \sqrt{V_0/2m}$

The estimate from our search gives exactly the correct ground state energy.
